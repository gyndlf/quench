```python
import yaml
from laboneq.dsl import LinearSweepParameter
import numpy as np
```

```python

start_freq = 300.0e6
stop_freq = 500.0e6
num_points = 401

freq_sweep = LinearSweepParameter(
    uid ="res_freq_set",
    start = start_freq,
    stop = stop_freq,
    count = num_points,
    axis_name="Frequency [Hz]"
)
```

```python
params = {
    "text": "this is a string...",
    "int": 3,
    "floaty": 3.23e-4,
    "dict": {
        "inside_value": 5235.32,
        "nested_dict": {
            "freq": freq_sweep
        },
        "unknown_class": np.linspace(0,1,10),
    }
}

params
        
```

```python
yaml.SafeDumper.yaml_representers[None] = lambda self, data: \
    yaml.representer.SafeRepresenter.represent_str(
        self,
        repr(data),
    )
```

```python
with open("test.yaml", "w") as yf:
    yf.write(yaml.safe_dump(params))
        
```

```python
with open("test.yaml", "r") as yf:
    e = yaml.safe_load(yf)
e
```
